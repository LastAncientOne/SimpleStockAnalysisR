# Simple Stock Plots in R

```r
library(quantmod)
library(TTR)
```

```r
getSymbols('AAPL', src = 'yahoo', from = as.Date('2019-01-01'), to = as.Date('2020-01-01'))
```

```r
AAPL <- na.omit(AAPL)
```

```r
head(AAPL)
```

```r
plot(Cl(AAPL), main="Line Chart of Apple", xlab="Date")
```

```r
barplot(Cl(AAPL), main="Bar Chart of Apple",
   xlab="Date")
```

```r
barplot(Cl(AAPL), main="Bar Chart of Apple", horiz=TRUE, ylab="Date")
```
